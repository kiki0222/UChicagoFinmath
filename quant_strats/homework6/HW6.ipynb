```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy as sp
import quandl
```

```python
#drop unnecessary tenor
tenor_map = {'THA':[int(x.split('-')[0]) for x in yc['THA'].columns][3:8],\
             'ROU':[1,3,5],\
             'JPN':[int(x.split('-')[0]) for x in yc['JPN'].columns][:5],\
             'IDN':[int(x.split('-')[0]) for x in yc['IDN'].columns][:5]}
yc['ROU'] = yc['ROU'].rename(columns = {'12-Month':'1-Year'})
```

```python
#transpose to prepare for trasition from spot rate to zero rate
spot_rates = {}
for c in currency:
    yc[c] = yc[c][[str(t)+'-Year' for t in tenor_map[c]]]
    df = yc[c].transpose().set_index(pd.Series(tenor_map[c]))
    spot_rates[c] = df*0.01
```

```python
#calculate zero rates
def compute_zcb_curve(spot_rates_curve):
    zcb_rates = spot_rates_curve.copy()
    for curve in spot_rates_curve.columns:
        spot = spot_rates_curve[curve]
        for tenor, spot_rate in spot.iteritems():
            if tenor>0.001:
                times = np.arange(tenor-0.5, 0, step=-0.5)[::-1]
                coupon_half_yr = 0.5*spot_rate
                z = np.interp(times, zcb_rates[curve].index.values, zcb_rates[curve].values) # Linear interpolation
                preceding_coupons_val = (coupon_half_yr*np.exp(-z*times)).sum()
                zcb_rates[curve][tenor] = -np.log((1-preceding_coupons_val)/(1+coupon_half_yr))/tenor
    return zcb_rates
```

```python
zero_rates = {}
for c in currency:
    zero_rates[c] = compute_zcb_curve(spot_rates[c])
```

```python
#look up the borrowing rate and decide which currency to borrow and which currency to borrow
libor['JPN'] = libor['JPN'].rename(columns = {'Yield':'Rate'})
for x in currency:
    libor[x] *= 0.1
    plt.plot(libor[x])
```

As we can see from the plot, YEN should be the borrowing currency and always have a position in the other three currency.

```python
#we start from the first wednesday after 2014-12-01.
import datetime
print(datetime.date(2014,12,1).weekday())
def getEveryDay(begin_date,end_date): 
    date_list = [] 
    begin_date = datetime.datetime.strptime(begin_date, "%Y-%m-%d") 
    end_date = datetime.datetime.strptime(end_date,"%Y-%m-%d") 
    while begin_date <= end_date: 
        date_str = begin_date.strftime("%Y-%m-%d") 
        date_list.append(pd.Timestamp(date_str)) 
        begin_date += datetime.timedelta(days=7) 
    return date_list 
begin = '2014-12-03'
end = '2020-04-30'
dates = getEveryDay(begin,end)
```

```python
#calculate cashflow
cash_flow = {}
for c in ['THA','ROU','IDN']:
    cash_flow[c] = pd.DataFrame(columns = {'borrow_cf','lend_cf','total_cf'})
```

```python
#get interest rate of shortened maturity
def get_ir(curve,td):
    z = np.interp(5-td, curve.index.values, curve.values)
    return z
```

```python
#for tha,rou,idn
for c in ['THA','ROU','IDN']:
    date_old = dates[0]    
    ir_old = zero_rates[c].loc[5,dates[0]]
    ir_old_jpn = libor['JPN'].loc[dates[0],'Rate']
    fx_old = fx[c].loc[dates[0],'RATE']
    fx_old_jpn = fx['JPN'].loc[dates[0],'RATE']
    cash_flow[c].loc[dates[0],'borrow_cf'] = 8
    cash_flow[c].loc[dates[0],'lend_cf'] = -10
    for i in range(1,len(dates)):
        date = dates[i]
        while (date in zero_rates[c].columns and date in libor['JPN'].index) == False:
            date += pd.Timedelta(days = 1)
        timedelta = (date-date_old).days/7/52
        
        ir = get_ir(zero_rates[c][date],timedelta)
        lend_side = 10*fx_old/(np.exp(-ir_old*5))*np.exp(-ir*(5-timedelta))/fx[c].loc[date,'RATE'] - 10
        borrow_side = -8*fx_old_jpn*(1+(ir_old_jpn+0.0001*50)*timedelta)/fx['JPN'].loc[date,'RATE'] + 8
        cash_flow[c].loc[date,'borrow_cf'] = borrow_side
        cash_flow[c].loc[date,'lend_cf'] = lend_side
        
        date_old = date
        ir_old = zero_rates[c].loc[5,date]
        ir_old_jpn = libor['JPN'].loc[date,'Rate']
        fx_old = fx[c].loc[date,'RATE']
        fx_old_jpn = fx['JPN'].loc[date,'RATE']
    cash_flow[c].loc[date,'lend_cf'] += 10
    cash_flow[c].loc[date,'borrow_cf'] -= 8
    cash_flow[c]['total_cf'] = cash_flow[c]['lend_cf'] + cash_flow[c]['borrow_cf']
    plt.figure()
    plt.plot(cash_flow[c]['total_cf'].cumsum())
    plt.title(c+' cash_flow_cumulative')
    plt.figure()
    plt.plot(cash_flow[c]['lend_cf'].cumsum())
    plt.title(c+' lending_cash_flow_cumulative')
    plt.figure()
    plt.plot(cash_flow[c]['borrow_cf'].cumsum())
    plt.title(c+' borrowing_cash_flow_cumulative')
```

Firstly, the jump of cash flow at the end of trading period is because we don't do the leverage trading anymore. We lend 10MM and borrow 8MM, so we'll have a net cash inflow of about 2MM at the end.
From the cash_flow_cumulative plots, we can see that except ROU, the other two fx carry trades generated an upward moving trend of cumulative cash flow, which means they did make profit during the period. Although we can tell from the plots that the volatility of cash flow is very large and the maximum drawdown also shows the uncertainty of loss in fx carry trades.
From the lending side cumulative cash flow plots, we can see that the profit generated by carry trades are mostly from this part. But both borrowing side cash flow and lending side cash flow contribute to the volatility of total cash flow for each foreign currency.

```python
#get the return
ret = {}
#I think we need to smooth the cash flow caused by 5/4 leverage
for c in ['THA','ROU','IDN']:
    cf = cash_flow[c][['total_cf']].copy()
    cf['total_cf'][0] += 2
    cf['total_cf'][-1] -= 2
    #plot adjusted total_cash_flow cumulative
    ret_df = pd.DataFrame(index = cf.index,columns = ['ret'])
    ret_df['ret'] = np.array(cf['total_cf'])/2.0
    ret[c] = ret_df
```

```python
c = 'THA'
print(("mean of "+c,ret[c].mean()))
print(("std of "+c,ret[c].std()))
print(("skew of "+c,ret[c].skew()))
print(("kurtosis of "+c,ret[c].kurtosis()))
print(("Sharpe of "+c,ret[c].mean()/ret[c].std()))
plt.figure()
plt.hist(ret[c]['ret'])
```

As we've already observed from the cash flow plots, only the mean returns of THA and IDN trades are positive. The annually approximate return of these two fx trades is about 0.2%*52 = 10.4% which is pretty high. The volatility of return for THA trade is lowest. 
For skewness, it's obvious that all three fx trades return shows negatively skewed characteristic of distribution, which is -0.2439,-0.4009 and -0.1081. This observation is very reasonable accroding to the intrinsic characteristic of carry trade. As we know, carry trade is like providing insurance for the interest rate risk and fx rate risk. During the whole trading period, we collect premium consistently. But once something unfavorable happens, the return can be wiped out and we are faced with extreme loss. We can also see the negative skewness from the histogram plots of returns from three fx carry trades.
For kurtosis, the kurtosis for THA, ROU, IDN are 3.465476, 5.092096 and 3.212889 respectively. So carry trade produced a fat tailed return, which means lots of extreme situations can happen doing carry trade. 
Finally, we can see that weekly sharpe of profiting trades are pretty good if we annulaized it. But that's probably the worst situation haven't happpened during the period we trade.
As a summary, I think we need to be very careful when generating trading strategies by looking into if they have return patterns or characteristics of carry trade.
