```python
import quandl
import pandas as pd
import datetime
import matplotlib.pyplot as plt

year = ['2018','2019']
month = ['H','M','U','Z']

tu = pd.DataFrame()
us = pd.DataFrame()
```

```python
#GET TU
last_day = pd.Timestamp(datetime.datetime(2017,12,1))
for y in year:
    for m in month:
        tu_name = "OWF/CBT_TU_TU_" + m + y + "_IVM"
        data_seg = quandl.get(tu_name, authtoken="khzKyrGRmxZen9BA5h9a")
        tu = tu.append(data_seg.loc[data_seg.index > last_day])
        last_day = data_seg.index[-1]
tu = tu.loc[tu.index <= pd.Timestamp(datetime.datetime(2019,8,31))]
tu
```

```python
#GET US
last_day = pd.Timestamp(datetime.datetime(2017,12,1))
for y in year:
    for m in month:
        us_name = "OWF/CBT_US_US_" + m + y + "_IVM"
        data_seg = quandl.get(us_name, authtoken="khzKyrGRmxZen9BA5h9a")
        us = us.append(data_seg.loc[data_seg.index > last_day])
        last_day = data_seg.index[-1]
us = us.loc[us.index <= pd.Timestamp(datetime.datetime(2019,8,31))]
us
```

```python
#SPREAD
spread = pd.DataFrame(us['Future']-tu['Future'])
spread = spread.dropna()
plt.plot(spread['Future'])
plt.xticks(rotation = 30)
```

```python
plt.hist(spread['Future'])
```

```python
print(spread.describe())
print(spread.skew())
print(spread.kurtosis())
```

```python
#DIFFERENCE
window = [5,10,20,40,80]
for w in window:
    d1 = spread - spread.rolling(w).mean()
    
    plt.figure()
    plt.plot(d1['Future'])
    plt.xticks(rotation = 30)
    plt.title(str(w)+'-day difference time series')
    
    plt.figure()
    plt.hist(d1['Future'])
    plt.title(str(w)+'-day difference histogram')
    
    print(d1.describe())
    print(d1.skew())
    print(d1.kurtosis())
```

```python
#DATA OUTPUT
data = pd.DataFrame({'spread':spread['Future'],'price_tu':tu['Future'],'price_us':us['Future']})
data = data.dropna()
data.to_csv('C:/Users/Krist/OneDrive/desktop/quant strategies/hw1/spread.csv')
```
